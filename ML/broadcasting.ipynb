# numpy broadcasting rules

## what broadcasting is

broadcasting is the mechanism numpy uses to perform operations on arrays of different shapes.

it allows numpy to automatically expand smaller arrays so they can match larger ones during operations.

---

## why broadcasting exists

without broadcasting you would need to manually reshape or copy arrays.

broadcasting makes operations efficient and avoids unnecessary memory usage.

---

## core rules of broadcasting

when operating on two arrays, numpy compares their shapes from right to left.

### rule 1

dimensions are compatible if they are equal  

or one of them is 1  

---

### rule 2

if one dimension is 1, it is stretched to match the other dimension  

this expansion is virtual and does not copy data in memory  

---

### rule 3

if dimensions do not match and neither is 1, broadcasting fails  

---

## step by step comparison

given two shapes:

A shape = (4, 3)  
B shape = (3,)  

align from right:

A = (4, 3)  
B = (1, 3)  

comparison:

3 matches 3  
1 expands to 4  

result shape becomes:

(4, 3)  

---
