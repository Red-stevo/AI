```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sn
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import random as rd
```

```python
dataset = load_iris(as_frame=True)
df_f = pd.DataFrame(dataset.data, columns=dataset.feature_names) 
df_f.head()
```

```python
df_t = pd.Series(dataset.target, name="Target")
df_t.head()
```

```python
df = df_f.join(df_t, how='left')
df.head()
```

```python
counts = df_t.value_counts()

counts.values.tolist()
```

```python
c_indexes = counts.index.tolist()
```

```python
def split_dataset(df, k:int, strat:bool, rand:bool):
    pair = []
    r_indexes = [i for i in range(len(df))]
    
    indexes = []
    for c in c_indexes:
        indexes.append(df[df == c].index.tolist())

    if strat:
        if rand:
            indexes = [np.random.permutation(i) for i in indexes]
        indicies = [np.array_split(arr, k) for arr in indexes]
        indexes = []
        for (arr1, arr2, arr3) in zip(*indicies):
            indexes.append(np.concat((arr1, arr2, arr3)))
    else:
        if rand:
            rd.shuffle(r_indexes)
        indexes = np.array_split(r_indexes, k)

    for i in range(len(indexes)):
        test = indexes[i]
        train = np.delete(indexes, i, axis=0)
        pair.append((test, train))
    return pair
```

```python
evaluation = {}
cfs = []
k_folds = 5
k_classes_min = 3
k_classes = 7
splits = split_dataset(df_t, k_folds, True, True)
for n in range(k_classes_min, k_classes):
    scores = []
    cf = []
    for tp in splits:
        model = KNeighborsClassifier(n_neighbors=n)
        for arr in tp[1]:
            model.fit(df_f.iloc[arr], df_t.iloc[arr])
        scores.append(model.score(df_f.iloc[tp[0]], df_t.iloc[tp[0]]))

        cf.append(confusion_matrix(df_t.iloc[tp[0]], model.predict(df_f.iloc[tp[0]])))
    evaluation[n] = float(np.mean(scores))
    cfs.append(cf)

evaluation
```

```python
len(cfs)
```

```python
fig, ax = plt.subplots((k_classes - k_classes_min), k_folds, figsize=(15, 10))


for i in range((k_classes - k_classes_min)):
    for h in range(k_folds):
        sn.heatmap(cfs[i][h], ax=ax[i][h], cmap='viridis', annot=True)
plt.tight_layout()        
```
