# Feature Selection and Dimensionality Reduction Techniques

Feature selection and dimensionality reduction are essential steps in machine learning to improve model performance, reduce overfitting, decrease training time, and enhance interpretability. These methods work by either **removing irrelevant features** or **retaining the most informative ones**.

Feature selection methods are commonly categorized into:

1. **Filter Methods**
2. **Wrapper Methods**
3. **Embedded Methods**

Additionally, **Principal Component Analysis (PCA)** is a **dimensionality reduction technique** that transforms features instead of selecting them.

---

# 1. Filter Techniques

Filter methods evaluate the **statistical relationship between each feature and the target variable independently of any machine learning model**. They are computationally efficient and often used as a **preprocessing step**.

They **remove features that have weak relationships with the target variable**.

## 1.a Correlation

### How It Works

Correlation measures the **linear relationship between two variables**. In feature selection, it is used to determine how strongly a feature is related to the target variable.

The most common measure is the **Pearson correlation coefficient**.

\[
r = \frac{cov(X,Y)}{\sigma_X \sigma_Y}
\]

Where:

- \( r \) ranges between **-1 and 1**
- **1** → strong positive relationship  
- **-1** → strong negative relationship  
- **0** → no linear relationship

### Removing and Retaining Features

**Retained Features**
- Features with **high absolute correlation with the target variable**

**Removed Features**
- Features with **very low correlation with the target**
- Highly correlated features with each other (**multicollinearity**) — one is removed to reduce redundancy.

Example:

| Feature | Correlation with Target |
|-------|------|
Age | 0.72 |
Income | 0.69 |
Zip Code | 0.01 |

Here **Zip Code would likely be removed**.

### Where It Is Applied

Commonly used in:

- **Linear Regression**
- **Logistic Regression**
- **Financial prediction models**
- **Preprocessing in tabular datasets**

It works best when **relationships are linear**.

---

## 1.b Mutual Information (MI)

### How It Works

Mutual Information measures the **amount of information one variable provides about another**. Unlike correlation, MI can capture **both linear and non-linear relationships**.

It is based on **information theory**.

\[
MI(X,Y) = H(X) + H(Y) - H(X,Y)
\]

Where:

- \(H(X)\) = entropy of variable X
- \(H(Y)\) = entropy of variable Y

Higher MI means the feature provides **more useful information about the target**.

### Removing and Retaining Features

**Retained Features**
- Features with **high mutual information scores**

**Removed Features**
- Features with **low MI scores** (little predictive value)

Example:

| Feature | MI Score |
|-------|------|
Clicks | 0.52 |
User Age | 0.41 |
Random ID | 0.00 |

Random ID would be removed.

### Where It Is Applied

Commonly used in:

- **Text classification**
- **Feature selection for NLP**
- **Categorical data**
- **Non-linear machine learning models**

Typical use cases include:

- Spam detection
- Document classification
- Recommendation systems

---

# 2. Wrapper Methods

Wrapper methods evaluate **subsets of features by actually training a machine learning model**. They search for the **best combination of features** based on model performance.

They are **more computationally expensive but often more accurate**.

---

## 2.a Forward Selection

### How It Works

Forward selection starts with **no features** and gradually adds features that improve the model the most.

Steps:

1. Start with **empty feature set**
2. Train models with **each individual feature**
3. Select the feature with **best performance**
4. Add it to the model
5. Repeat by testing remaining features
6. Stop when no further improvement occurs

Example progression:

| Step | Selected Features |
|----|----|
1 | Age |
2 | Age, Income |
3 | Age, Income, Education |

### Removing and Retaining Features

**Retained Features**
- Features that **increase model performance**

**Removed Features**
- Features that **do not significantly improve accuracy**

### Where It Is Applied

Used when:

- Dataset has **moderate number of features**
- Feature interactions are important

Common models:

- Logistic Regression
- Linear Regression
- Support Vector Machines

---

## 2.b Backward Selection

### How It Works

Backward selection starts with **all features** and progressively removes the least useful ones.

Steps:

1. Start with **all features**
2. Train the model
3. Remove the **least important feature**
4. Retrain the model
5. Continue until performance degrades

Example progression:

| Step | Remaining Features |
|----|----|
1 | Age, Income, Education, Zip |
2 | Age, Income, Education |
3 | Age, Income |

### Removing and Retaining Features

**Removed Features**
- Features whose removal **does not hurt performance**

**Retained Features**
- Features necessary for maintaining **model accuracy**

### Where It Is Applied

Common in:

- Statistical modeling
- Econometrics
- Medical research
- Regression analysis

---

# 3. Embedded Methods

Embedded methods perform **feature selection during the training of the machine learning model itself**.

The algorithm **automatically learns which features are important**.

---

## 3.a Lasso Regression

### How It Works

Lasso (Least Absolute Shrinkage and Selection Operator) adds **L1 regularization** to linear regression.

Loss function:

\[
Loss = RSS + \lambda \sum |w_i|
\]

Where:

- \(RSS\) = residual sum of squares
- \(w_i\) = model coefficients
- \(\lambda\) = regularization parameter

L1 regularization forces **some coefficients to become exactly zero**.

### Removing and Retaining Features

**Retained Features**
- Features with **non-zero coefficients**

**Removed Features**
- Features whose coefficients shrink to **zero**

Example:

| Feature | Coefficient |
|------|------|
Age | 0.45 |
Income | 0.31 |
Zip Code | 0 |

Zip Code is removed.

### Where It Is Applied

Common in:

- **High dimensional datasets**
- **Genomics**
- **Finance**
- **Sparse models**

---

## 3.b Tree-Based Models

### How It Works

Tree-based algorithms determine feature importance based on **how much each feature reduces impurity** when splitting the data.

Common impurity measures:

- **Gini Index**
- **Entropy**
- **Variance reduction**

Features that **create better splits** are considered more important.

Example:

A feature that frequently splits nodes near the top of the tree is considered highly important.

### Removing and Retaining Features

**Retained Features**
- Features with **high feature importance scores**

**Removed Features**
- Features rarely used for splitting

### Where It Is Applied

Used in models such as:

- Random Forest
- Gradient Boosting
- XGBoost
- Decision Trees

These models are widely applied in:

- Fraud detection
- Customer churn prediction
- Credit scoring
- Medical diagnosis

---

# Principal Component Analysis (PCA)

PCA is a **dimensionality reduction technique**, not strictly a feature selection method.

Instead of selecting features, it **transforms the original features into new orthogonal variables called principal components**.

---

## How PCA Works

PCA identifies directions (components) in which the **variance of the data is maximized**.

Steps:

1. Standardize the dataset
2. Compute the **covariance matrix**
3. Calculate **eigenvectors and eigenvalues**
4. Rank components by **explained variance**
5. Select the top components
---

## Removing and Retaining Features

**Retained Information**
- Principal components explaining the **largest variance**

**Removed Information**
- Components explaining **small variance**

Example:

| Component | Variance Explained |
|----|----|
PC1 | 60% |
PC2 | 25% |
PC3 | 5% |

PC3 may be discarded.

---

## Where PCA Is Applied

Commonly used in:

- **Computer vision**
- **Face recognition**
- **Image compression**
- **Noise reduction**
- **Data visualization**

Also used before training models when datasets contain **hundreds or thousands of features**.

---

# Summary Comparison

| Method | Type | Model Required | Main Idea |
|------|------|------|------|
Correlation | Filter | No | Remove weakly correlated features |
Mutual Information | Filter | No | Measure information shared with target |
Forward Selection | Wrapper | Yes | Add features progressively |
Backward Selection | Wrapper | Yes | Remove features progressively |
Lasso Regression | Embedded | Yes | Shrink coefficients to zero |
Tree-Based Models | Embedded | Yes | Select features based on split importance |
PCA | Dimensionality Reduction | No | Transform features into principal components |
