# Understanding numpy linspace

## What linspace does

numpy linspace generates a sequence of evenly spaced numbers between a start value and an end value.

You define:

* start value
* end value
* number of points

numpy computes values that are equally spaced across that interval.

---

## Basic example

```python
import numpy as np

np.linspace(0, 10, 5)
```

**Output**

```python
[0.  2.5  5.  7.5 10.]
```

This creates 5 values starting at 0 and ending at 10.

---

## How it works internally

The spacing between values is calculated as:

step = (end minus start) divided by (number of points minus 1)

### Example

* start = 0
* end = 10
* points = 5

step = (10 minus 0) divided by (5 minus 1)
step = 10 divided by 4
step = 2.5

### Generated values

* 0
* 0 plus 2.5
* 0 plus 2.5 times 2
* 0 plus 2.5 times 3
* 0 plus 2.5 times 4

---

## Including or excluding the end value

By default, the end value is included.

```python
np.linspace(0, 10, 5)
```

### Excluding the end value

```python
np.linspace(0, 10, 5, endpoint=False)
```

This changes the spacing so the values stop before 10.

---

## Returning the step size

You can also get the step value:

```python
values, step = np.linspace(0, 10, 5, retstep=True)
```

### Output

**values**

```python
[0.  2.5  5.  7.5 10.]
```

**step**

```python
2.5
```

---

## Data type control

You can control the output type:

```python
np.linspace(0, 10, 5, dtype=int)
```

---

## linspace vs arange

### linspace

* you control number of values

### arange

* you control step size

### Example

```python
np.linspace(0, 10, 5)
```

```python
np.arange(0, 10, 2)
```

---

## Summary

* linspace splits a range into equal parts based on how many values you want
* it guarantees inclusion of the end value unless changed
* it is ideal when you need a fixed number of evenly spaced samples


````python
np.linspace(1, 6, 6)
````

````python
np.linspace(1, 50)
````

````python
np.linspace(-1, 1, 20)
````

````python
np.linspace(1, -1, 20)
````

# Understanding numpy arange

## What arange does

numpy arange generates a sequence of numbers using a fixed step size.

You define:

* start value
* end value
* step size

numpy then builds values by repeatedly adding the step until it reaches the end.

---

## Basic example

```python id="p1x8sa"
import numpy as np

np.arange(0, 10, 2)
```

**Output**

```python id="k9v2lm"
[0 2 4 6 8]
```

This creates values starting at 0 and increases by 2 each time.

---

## How it works internally

The sequence is built like this:

value = start

next value = value plus step

repeat until the value is greater than or equal to the end

### Example

* start = 0
* end = 10
* step = 2

### Generated values

* 0
* 2
* 4
* 6
* 8

10 is not included because the sequence stops before reaching the end.

---

## End value behavior

The end value is not included.

```python id="r8gqxo"
np.arange(0, 10, 2)
```

Even though 10 fits the pattern, it is excluded because arange stops before the end.

---

## Floating point steps

arange can use decimal steps, but this can introduce precision issues.

```python id="a0m9tq"
np.arange(0, 1, 0.2)
```

Possible output:

```python id="6q4r8n"
[0.  0.2 0.4 0.6 0.8]
```

Sometimes the final value may not be exact due to floating point rounding.

---

## Data type control

You can control the output type:

```python id="d7n2yk"
np.arange(0, 10, 2, dtype=float)
```

---

## When to use arange

Use arange when you care about the step size between values.

---

## arange vs linspace

### arange

* you control step size
* end is usually excluded
* can have floating point precision issues

### linspace

* you control number of values
* end is included by default
* more stable for decimal spacing

### Example

```python id="l1z8pd"
np.arange(0, 10, 2)
```

```python id="c5h3fw"
np.linspace(0, 10, 5)
```

---

## Summary

* arange builds sequences using a fixed step size
* it stops before reaching the end value
* it is best when you need consistent increments between values
* be careful when using decimal step sizes due to precision limitations


````python
np.arange(1,0)
````

````python
np.arange(1, 50, 60)
````

````python
np.arange(0, 50, 50)
````

````python
np.arange(0, 200, 5)
````

````python
np.arange(61)
````

# Understanding numpy reshape

## What reshape does

numpy reshape changes the shape of an array without changing its data.

It reorganizes how the data is viewed in dimensions.

You define:

* the original array
* the new shape

The total number of elements must remain the same.

---

## Basic example

```python
import numpy as np

arr = np.arange(6)
reshaped = arr.reshape(2, 3)

print(reshaped)
```

**Output**

```python
[[0 1 2]
 [3 4 5]]
```

The data is the same, only arranged into 2 rows and 3 columns.

---

## Key rule

number of elements must match

Example:

* original array has 6 elements
* new shape must also contain 6 elements

Valid shapes:

* 2 by 3
* 3 by 2
* 1 by 6
* 6 by 1

Invalid:

* 4 by 2  because it requires 8 elements

---

## How reshape works internally

reshape reads the array in row order and fills the new shape in the same order.

Original:

```python
[0 1 2 3 4 5]
```

Reshaped to 2 by 3:

```python
[[0 1 2]
 [3 4 5]]
```

---

## Using negative one

You can let numpy infer one dimension using negative one.

```python
arr = np.arange(6)

reshaped = arr.reshape(2, -1)
print(reshaped)
```

**Output**

```python
[[0 1 2]
 [3 4 5]]
```

numpy automatically calculates the missing dimension.

---

## Flattening an array

Convert multi dimensional array back to one dimension:

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

flat = arr.reshape(-1)
print(flat)
```

**Output**

```python
[1 2 3 4 5 6]
```

---

## Reshaping to higher dimensions

```python
arr = np.arange(8)

reshaped = arr.reshape(2, 2, 2)
print(reshaped)
```

**Output**

```python
[[[0 1]
  [2 3]]

 [[4 5]
  [6 7]]]
```

---

## Common patterns

Row vector

```python
arr.reshape(1, -1)
```

Column vector

```python
arr.reshape(-1, 1)
```

---

## reshape vs resize

reshape:

* returns a new view or array
* does not modify original shape

resize:

* modifies the original array
* can change total number of elements

---

## Summary

* reshape changes how data is arranged, not the data itself
* total number of elements must remain constant
* negative one allows automatic dimension calculation
* useful for preparing data for models and matrix operations

````python
arr.reshape(4, -1)
````

# Understanding numpy resize

## What resize does

numpy resize changes the shape of an array and can also change the number of elements.

Unlike reshape, resize can:

* modify the original array
* increase or decrease the total number of elements

If the new size is larger, numpy repeats the original data.

---

## Basic example

```python
import numpy as np

arr = np.array([1, 2, 3, 4])

np.resize(arr, (2, 3))
```

**Output**

```python
[[1 2 3]
 [4 1 2]]
```

The array is expanded to 6 elements, so values repeat.

---

## Increasing size

```python
arr = np.array([1, 2, 3])

resized = np.resize(arr, 7)
print(resized)
```

**Output**

```python
[1 2 3 1 2 3 1]
```

The data repeats to fill the new size.

---

## Decreasing size

```python
arr = np.array([1, 2, 3, 4, 5])

resized = np.resize(arr, 3)
print(resized)
```

**Output**

```python
[1 2 3]
```

Extra elements are discarded.

---

## Important behavior

resize does not preserve structure

It simply fills the new shape in order, repeating data if needed.

---

## In place resize method

There is also an in place version:

```python
arr = np.array([1, 2, 3, 4])

arr.resize((2, 2))
print(arr)
```

**Output**

```python
[[1 2]
 [3 4]]
```

This modifies the original array directly.

---

## Key differences from reshape

reshape:

* keeps total number of elements the same
* does not repeat or remove data
* usually returns a new array

resize:

* can change total number of elements
* repeats or truncates data
* can modify the original array

---

## When to use resize

Use resize when:

* you need a specific shape regardless of original size
* repeating data is acceptable

Avoid resize when:

* data integrity is important
* you do not want duplicated values

---

## Summary

* resize changes both shape and size
* it can repeat or remove data
* it can modify the original array
* it is less strict than reshape and should be used carefully
