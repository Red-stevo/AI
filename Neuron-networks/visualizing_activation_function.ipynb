```python
import numpy as np
from matplotlib import pyplot as plt
```

# Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
sigmoid(-2)
```

```python
vals = np.linspace(-100, 100, 100000)
vals.shape
```

```python
y = sigmoid(vals)
y
```

```python
plt.plot(vals, y)
```

```python
x = np.linspace(-10, 10, 100000000)
plt.plot(x, sigmoid(x))
```
