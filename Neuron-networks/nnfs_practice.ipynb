# nnfs data generation methods

in the Neural Networks from Scratch approach, a few helper functions are heavily used to generate synthetic datasets for testing models.

the most important ones are:

* spiral data
* vertical data
* sine data

---

# spiral data generation

## purpose

creates a dataset where classes form spiral patterns.

this is useful for testing non linear models because the data is not linearly separable.

---

## method

```python
from nnfs.datasets import spiral_data

X, y = spiral_data(samples=100, classes=3)
```

---

## what it generates

X shape

number of samples times number of classes by 2

example

100 samples and 3 classes gives 300 by 2

y shape

300 values representing class labels

---

## how it works conceptually

for each class:

* generate radius values increasing outward
* generate angle values with some noise
* convert polar coordinates to cartesian

so each class forms a spiral arm.

---

## why it is important

* tests ability of networks to learn complex boundaries
* widely used in nnfs examples
* shows failure of linear classifiers clearly

---

# vertical data generation

## purpose

creates simple linearly separable data.

used for testing basic models and debugging.

---

## method

```python
from nnfs.datasets import vertical_data

X, y = vertical_data(samples=100, classes=3)
```

---

## what it generates

data points arranged in vertical stripes.

each class occupies a different region along the x axis.

---

## behavior

* easy for linear models
* useful for verifying correctness of implementation

---

# sine data generation

## purpose

used for regression tasks.

models try to learn a continuous function.

---

## method

```python
from nnfs.datasets import sine_data

X, y = sine_data()
```

---

## what it generates

X contains input values along a range

y contains sine of those values

---

## behavior

* tests regression capability
* useful for understanding loss functions like mean squared error

---

# key differences

spiral data

* non linear classification
* hard problem

vertical data

* linear classification
* easy problem

sine data

* regression problem
* continuous output

---

# summary

these dataset generators are used to:

* test model behavior under different complexity
* visualize learning
* debug implementations

spiral data is the most important for classification

vertical data is best for simple checks

sine data is used for regression tasks
