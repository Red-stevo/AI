```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences = [
    "the cat sat on the mat",
    "the dog sat on the log", 
    "cats and dogs are pets"
]

conveter = CountVectorizer()
```

```python
transformed_sentence = conveter.fit_transform(sentences)
transformed_sentence.toarray()
```

```python
conveter.vocabulary_
```

```python
conveter.get_feature_names_out()
```

```python
vectorizer = CountVectorizer(stop_words='english')
```

```python
transformed_sentence = vectorizer.fit_transform(sentences)
```

```python
vectorizer.get_feature_names_out()
```

```python
transformed_sentence.toarray()
```
